```python
import os
import pandas as pd

data = []  # List to store data

for i in range(1, 14):
    folder_path = '/content/drive/MyDrive/ProjectData/s' + str(i) + '/txt/'
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, "r") as file:
            text = file.read()
            data.append({'Path': file_path, 'Text': text})

df = pd.DataFrame(data)  # Create a dataframe from the list of dictionaries
print(df)
```

```python
import spacy
import pandas as pd

# Load the spacy model
nlp = spacy.load("en_core_web_lg")

# Create a new dataframe to store the combined information
combined_df = pd.DataFrame(columns=["Path", "Text", "NER", "INDEXING"])

for index, row in df.iterrows():
    # Get the text from the original dataframe
    text = row["Text"]

    # Apply NER on the text
    doc = nlp(text)

    # Extract NER information
    ner_data = []
    indexi = []
    for entity in doc.ents:
        ner_data.append(entity.text + "(" + entity.label_ + ")")
        indexi.append(entity.label_ + ":" + str(entity.start))

    # Create a new row for the combined dataframe
    combined_row = {
        "Path": row["Path"],
        "Text": text,
        "NER": ", ".join(ner_data),
        "INDEXING": indexi
    }

    # Append the row to the combined dataframe
    combined_df = combined_df.append(combined_row, ignore_index=True)

print(combined_df)
```

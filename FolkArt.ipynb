```python
import torch
from PIL import Image
# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)


model.conf = 0.04 # Lower threshold for more detections

# Images
image = Image.open("/content/drive/MyDrive/ProjectData/s1/img/s10.jpg")
imgs = image.convert("L")
# Inference
results = model(imgs)

# Show results on image with bounding boxes
results.show()
results.print()
results.save()

results.xyxy[0]
results.pandas().xyxy[0]
```
